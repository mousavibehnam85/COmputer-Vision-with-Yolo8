```python
from ultralytics import YOLO


# Load the pre-trained YOLOv8 model
model = YOLO('yolov8n.pt')  # Ensure 'yolov8n.pt' is the correct path to your model

# Specify the path to your image
image_path = 'FighterJets1.png'  # Update this path to your image


model.conf = 0.25 
# Perform detection on the image
results = model(image_path)


#if results:
#    print(type(results[0]))
#    print(results[0])



# Accessing predictions for the first image (assuming single image inference)
# Each prediction includes (x1, y1, x2, y2, confidence, class_id)
detections = results[0]

print(type(results))

```
